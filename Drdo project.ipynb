```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
print(housing.data.shape)
print(housing.feature_names)

```

## Preparing the Dataset

```python
dataset=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=housing.target
```

```python
## Independent and dependent
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
##Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
##Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
## print the coffiecent and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameter the model trained
regression.get_params()
```

```python
## Predection on the test data
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scatter plot from the predication
plt.scatter(y_test,reg_pred)
```

```python
## Residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## plot the residuals
sns.displot(residuals,kind="kde")
```

```python
##Scatter plot with respect residuals and prediction
##uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

## R sqaure  and Adjusted r square
Formula:
R^2=1-SSR/SST
R^2ua=Cofficient of detrmination
SSR=sum of sqaure of residuals
SST=total sum of square

```python

from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

##Adjust R2=1-[(1-R2)*(n-1)/(n-k-1)]
Where;
R2=The R2 of model
n= No. of observation
k=no.of predicator variable

```python
#display adjucted R sqaured
1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
